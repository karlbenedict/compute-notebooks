```python
import urllib.request
import datetime
import json
```

```python
# Get the package metadata from Data.gov
requestSize = 1
requestLimit = 5
requestBaseURL = "https://catalog.data.gov/api/3/action"
requestFunction = "package_search"
saveDirectory = "Data.gov_Files"

i=0
finished = False
timestamp = '{:%Y-%m-%d-%H:%M:%S}'.format(datetime.datetime.now())
numDataSets = 0

while not finished:
    requestString = requestBaseURL + "/" + requestFunction + "?" + "rows=" + str(requestSize) + "&" + "start=" + str(i*requestSize)
    print(str(i) + ".\tgetting: " + requestString)
    
    outFilePath = saveDirectory + "/" + "DataGov_" + timestamp + "_" + '{0:08d}'.format((i*requestSize)+1) + "-" + '{0:08d}'.format((i+1)*requestSize) + ".json"
    print("\toutput file path: " + outFilePath)
    
    with urllib.request.urlopen(requestString) as JSONfile:    
        resultString = JSONfile.read().decode('utf-8')
        #print(resultString) 
        numDataSets = json.loads(resultString)['result']['count']
        with open(outFilePath, "w") as f:
            f.write(resultString)
    
    i += 1
    print("\t" + str(i*requestSize) + " out of " + str(numDataSets) + " processed")
    if i >= requestLimit: finished = True
    if i*requestSize > numDataSets: finished = True 
    
```
