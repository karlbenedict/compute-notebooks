```python
# Load dependencies.

from wordcloud import WordCloud, STOPWORDS
from scipy.misc import imread
import docx
import codecs
```

## From a Word File

```python
# Replace the empty string with the name of your Word file.
# To avoid errors, make sure the Word file is in the same directory as this notebook!

doc = docx.Document('darwin_origin.docx')
fullText = []
for para in doc.paragraphs:
    fullText.append(para.text)
```

```python
# Extract paragraph text.

corpus = ''
for p in fullText:
    corpus = corpus + str(p)
    
#corpus
```

```python
# Generate the word cloud!

wordcloud = WordCloud(max_words=50,
                      min_font_size=6,
                      background_color='#ffffff',
                      mask=imread('cloud_mask_1.png')
                     )
wordcloud.generate_from_text(corpus)
wordcloud.to_image()
```

## From a Text File

```python
# Replace the given file name with yours.
import string

wc = WordCloud(max_words=50,
                      min_font_size=6,
                      background_color='#ffffff',
                      mask=imread('cloud_mask_1.png')
                     )

text = []
with open('darwin_origin.txt', 'r') as data:
    for line in data.readlines():
        d = codecs.encode(line, 'utf-8')
        text.append(d)
    data.close()
    
c = ''
for line in text:
    c = c + str(line)
    
c = c.replace("\\n\'b\'", '')
wc.generate_from_text(c)

wc.to_image()
```
